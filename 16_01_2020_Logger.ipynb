```python
import logging
logging.basicConfig(filename =  'test2.log', level = logging.DEBUG, format = '%(levelname)s %(asctime)s %(message)s')
# time and level are accounted here

logging.info('This is my info log')
logging.warning('This is my warning log')
logging.debug('This is my debug log')
logging.error('This is my error log')
```

# Priority level
1. ERROR
2. WARNING
3. INFO
4. DEBUG

If we set ERROR as level then log wont contain WARING, INFO, and DEBUG logs as Error has high priority and it wont come down to debug. On the other hand if we set DEBUG as priority level it stores all the remaining logs.


```python
import logging
logging.basicConfig(filename =  'test3.log', level = logging.INFO, format = '%(levelname)s %(asctime)s %(message)s')
# time and level are accounted here

logging.info('This is my info log')
logging.warning('This is my warning log')
logging.debug('This is my debug log')
logging.error('This is my error log')
```

```python
def divbyzero(a,b):
    logging.info('This is a start of my code and I am trying to enter %s and %s', a,b)
    try:
        result = a/b
        logging.info('Executed Successfully')
        
    except Exception as e:
        logging.error('Error has occured')
        logging.exception('Exception occured: ' + str(e))        
```
